```python
titles = pd.read_csv('data/titles.csv')
titles = titles.dropna()
titles.head()
```

```python
cast = pd.read_csv('data/cast.csv')
cast = cast.dropna()
cast.head()
```

### What are the ten most common movie names of all time?

```python
t = titles
t.title.value_counts().head(10)
```

### Which three years of the 1930s saw the most films released?

```python
t = titles
t = titles[(titles.year >= 1930) & (titles.year < 1940)]
t.year.value_counts()
```

### Plot the number of films that have been released each decade over the history of cinema.

```python
t = titles
t.year = (t.year // 10) * 10
t = t.year.value_counts().sort_index()
t.plot(kind = "bar")
# t = titles.year.value_counts().sort_index()
# t.plot()
```

### Plot the number of "Hamlet" films made each decade.

```python
t = titles
t = t[t.title.notnull()]
t = t[t.title.str.startswith("Hamlet")]
t.year = t.year // 10
t.year *= 10
t.year.value_counts().sort_index().plot(x="year")
```

### Plot the number of "Rustler" characters in each decade of the history of film.

```python
c = cast
c = c[c["character"].str.startswith("Rustler")]
c = c.year // 10
c = c * 10
c = c.value_counts()
c.sort_index().plot(x="year")
```

### Plot the number of "Hamlet" characters each decade.
